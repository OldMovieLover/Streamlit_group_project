```python
df = df.drop(['Телефоны', 'Название ЖК', 'Серия дома'], axis=1)
```

Разобъем Количество комнат на число комнат и Тип смежности комнат.

```python
df[['Число комнат', 'Тип смежности']] = df['Количество комнат'].str.extract(r'(\d+),?\s*(.*)')
room_columns = list(df.columns)
room_columns.insert(room_columns.index('Количество комнат') + 1, room_columns.pop(room_columns.index('Число комнат')))
room_columns.insert(room_columns.index('Количество комнат') + 2, room_columns.pop(room_columns.index('Тип смежности')))
df = df[room_columns]
```

```python
df['Число комнат'] = pd.to_numeric(df['Число комнат'])
```

```python
df.head()
```

Преобразуем площадь в общую площадь

```python
df['Общая площадь, м2'] = df['Площадь, м2'].str.split('/').str[0].astype(float)
columns = list(df.columns)
columns.insert(columns.index('Площадь, м2') + 1, columns.pop(columns.index('Общая площадь, м2')))
df = df[columns]
df.head()
```

Разобъем колонку Дом на Этаж и Тип дома.

```python
df[['Этаж', 'Этажность', 'Тип дома']] = df['Дом'].str.extract(r'(\d+)/(\d+),?\s*(.*)')
flor_columns = list(df.columns)
flor_columns.insert(flor_columns.index('Дом') + 1, flor_columns.pop(flor_columns.index('Этаж')))
flor_columns.insert(flor_columns.index('Дом') + 2, flor_columns.pop(flor_columns.index('Этажность')))
flor_columns.insert(flor_columns.index('Дом') + 3, flor_columns.pop(flor_columns.index('Тип дома')))
df = df[flor_columns]
df['Этаж'] = pd.to_numeric(df['Этаж'])
df
```

```python
MissingValue = df.isnull().sum().sort_values(ascending = False)
Percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
MissingData = pd.concat([MissingValue, Percent], axis=1, keys=['Пропущенные значения', 'Процент'])
MissingData
```
