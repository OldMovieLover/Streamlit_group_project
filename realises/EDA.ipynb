# Расмотрим количество не заполненых строк.

```python
df.isnull().sum()
```

```python
MissingValue = df.isnull().sum().sort_values(ascending = False)
Percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
MissingData = pd.concat([MissingValue, Percent], axis=1, keys=['Пропущенные значения', 'Процент'])
MissingData
```

```python
colours = ['#993366', '#FFFF00']
sns.heatmap(df.isnull(), cmap=sns.color_palette(colours))
# Decorations
plt.title('Матрица пропущенных значений набора данных', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

# Посмотрим сколько комнат чаше сдают в аренду.

```python
df_prov = df['Количество комнат'].value_counts().to_frame().head(10)
df_prov.reset_index(inplace=True)
df_prov.columns = ['Количество комнат','counts']
fig, ax = plt.subplots(figsize=(10,6));
sns.barplot(data = df_prov, x='Количество комнат',y='counts',ax=ax).set_title('Колличество комнат');
plt.xticks(rotation='vertical');
```

# Расмотрим какую квадратуру чаще всего сдают.

```python
area_room = df['Площадь, м2'].value_counts().head(10)
area_room.plot(kind='pie', autopct='%1.1f%%')
plt.title('Распределение типов недвижимости')
plt.ylabel('')
plt.show()
```

# Построим диаграму рассеяния: Цена vs Высота потолков.

```python
top_jk = df['Цена'].value_counts().nlargest(10).index
filtered_df = df[df['Цена'].isin(top_jk)]
fig, ax = plt.subplots(figsize=(10, 6))
filtered_high_df = filtered_df[filtered_df['Высота потолков, м'] < 10]
plt.scatter(filtered_high_df['Высота потолков, м'], filtered_high_df['Цена'], alpha=0.5)
plt.yticks(filtered_high_df['Цена'], [str(value)[:8] for value in filtered_high_df['Цена']])
plt.title('Диаграмма рассеяния: Цена vs Высота потолков')
plt.ylabel('Цена')
plt.xlabel('Высота потолков, м')
plt.grid(True)
plt.tight_layout()
plt.show()
```
