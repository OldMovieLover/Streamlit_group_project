## Разобъем Количество комнат на число комнат и Тип смежности комнат.

```python
df[['rooms_count', 'type_rooms']] = df['counts_room'].str.extract(r'(\d+),?\s*(.*)')
room_columns = list(df.columns)
room_columns.insert(room_columns.index('counts_room') + 1, room_columns.pop(room_columns.index('rooms_count')))
room_columns.insert(room_columns.index('counts_room') + 2, room_columns.pop(room_columns.index('type_rooms')))
df = df[room_columns]
df['rooms_count'] = pd.to_numeric(df['rooms_count'])
```

## Разобъем на станцию метро и удаленность от метро.

```python
df[['metro_station', 'time_to_metro']] = df['metro'].str.extract(r'м\.\s*([^\(]+)\s*\((\d+)\s*мин')
```

## Разобъем на регион-город и улицу.

```python
df[['city', 'address']] = df['adress'].str.extract(r'(?P<Город>[А-яёЁ\- ]+),\s*(?P<Улица>.+)')
adres_columns = list(df.columns)
adres_columns.insert(adres_columns.index('adress') + 1, adres_columns.pop(adres_columns.index('city')))
adres_columns.insert(adres_columns.index('adress') + 2, adres_columns.pop(adres_columns.index('address')))
df = df[adres_columns]
```

## Разобъем на общую площадь, жилую и кухню.

```python
def split_squares(s):
    parts = s.split('/')
    if len(parts) == 2:
        return [parts[0], None, parts[1]]
    elif len(parts) == 3:
        return parts
    else:
        return [s, None, None]
df[['all_squar', 'live_squar', 'kitchen_squar']] = df['squar'].apply(split_squares).apply(pd.Series)
```

## Разобъем на этаж, этажность и тип дома.

```python
df[['floor', 'all_floor', 'type_apart']] = df['house_info'].str.extract(r'(\d+)/(\d+),?\s*(.*)')
flor_columns = list(df.columns)
flor_columns.insert(flor_columns.index('house_info') + 1, flor_columns.pop(flor_columns.index('floor')))
flor_columns.insert(flor_columns.index('house_info') + 2, flor_columns.pop(flor_columns.index('all_floor')))
flor_columns.insert(flor_columns.index('house_info') + 3, flor_columns.pop(flor_columns.index('type_apart')))
df = df[flor_columns]
df['floor'] = pd.to_numeric(df['floor'])
```

## Разобъем на цену за месяц, коммунальные услуги, залог, срок аренды и предоплату.

```python
list_el_price = [el for el in df['price'].str.split(',')]
df['ptice_month'] = [int(el[0].split()[0].split('.')[0]) for el in list_el_price]
df['util_service'] = df['price'].str.contains('Коммунальные услуги включены').astype(int)
df['bail'] = df['price'].str.extract(r'Залог - (\d+)')[0].fillna(0)
time_rent = df['price'].str.extract(r'Срок аренды - (.*?)(?:,|$)')[0]
df['rent_period'] = time_rent.apply(lambda x: 1 if 'Длительный' in x else 0)
df['prepayment'] = df['price'].str.extract(r'Предоплата (\d+) мес')[0].fillna(0)
```

## Разобъем на пассажирский и грузовой лифт.

```python
df['pas_lift'] = df['lift'].str.extract(r'Пасс \((\d+)\)').fillna(0).astype(int)
df['freight_lift'] = df['lift'].str.extract(r'Груз \((\d+)\)').fillna(0).astype(int)
```

## Разобъем на можно с детьми, животными и оба варианта.

```python
df['сhild_allow'] = df['child_animal'].apply(lambda x: 'да' if x == 'Можно с детьми' in str(x) else 'нет')
df['pets_allow'] = df['child_animal'].apply(lambda x: 'да' if x == 'Можно с животными' in str(x) else 'нет')
df['child_pets_allow'] = df['child_animal'].apply(lambda x: 'да' if x == 'Можно с детьми, Можно с животными' in str(x) else 'нет')
df['child_pets_allow'] = df['child_animal'].apply(lambda x: 'да' if pd.isna(x) else 'нет')
```

## Разобъем на количество санузлов и тип смежности.

```python
def extract_total_bathrooms(value):
    if pd.isna(value):
        return np.nan
    else:
        total = sum(map(int, [x.split('(')[1].strip(')') for x in value.split(', ')]))
        return total

def extract_bathroom_type(value):
    if pd.isna(value):
        return np.nan
    elif 'Совмещенный' in value and 'Раздельный' in value:
        return 'Совмещенный и Раздельный'
    elif 'Совмещенный' in value:
        return 'Совмещенный'
    elif 'Раздельный' in value:
        return 'Раздельный'

df['bathrooms'] = df['wc'].apply(extract_total_bathrooms)
df['type_bathrooms'] = df['wc'].apply(extract_bathroom_type)
```

# Заполняем пропуски.

### Если адрес один, то и высота потолков одна.

```python
for address in df['address'].unique():
    heights = df.loc[df['address'] == address, 'ceil'].dropna()
    if not heights.empty:
        df.loc[(df['address'] == address) & (df['ceil'].isnull()), 'ceil'] = heights.iloc[0]
```

### Если адрес один, то и мусоропровод присутствует.

```python
for address in df['address'].unique():
    heights = df.loc[df['address'] == address, 'trash'].dropna()
    if not heights.empty:
        df.loc[(df['address'] == address) & (df['trash'].isnull()), 'trash'] = heights.iloc[0]
```

```python
df['trash'] = df['trash'].fillna('Нет')
```

### Если адрес один, то метро ближайщее одно.

```python
for address in df['address'].unique():
    heights = df.loc[df['address'] == address, 'metro_station'].dropna()
    if not heights.empty:
        df.loc[(df['address'] == address) & (df['metro_station'].isnull()), 'metro_station'] = heights.iloc[0]
```

```python
df['metro_station'] = df['metro_station'].fillna('Нет')
```

### Если адрес один, то и приблизительное время до метро одно.

```python
for address in df['address'].unique():
    heights = df.loc[df['address'] == address, 'time_to_metro'].dropna()
    if not heights.empty:
        df.loc[(df['address'] == address) & (df['time_to_metro'].isnull()), 'time_to_metro'] = heights.iloc[0]
```

### Если адрес один и этаж не первый, то балкон присутствует. Остальное добавим по моде.

```python
for address in df['address'].unique():
    floors = df.loc[df['address'] == address, 'floor'].unique()
    
    for floor in floors:
        heights = df.loc[(df['address'] == address) & (df['floor'] == floor), 'balcon'].dropna()
        if not heights.empty:
            df.loc[(df['address'] == address) & (df['floor'] != 1) & (df['balcon'].isnull()), 'balcon'] = heights.iloc[0]
```

```python
balcon_mode = df['balcon'].mode()[0]
df['balcon'] = df['balcon'].fillna(balcon_mode)
```

### Комментарий от Артемия это его код)))

```python
df['all_floor'] = pd.to_numeric(df['all_floor'], errors='coerce')
df['rooms_count'] = pd.to_numeric(df['rooms_count'], errors='coerce')



def fill_room_count(row):
    if pd.isna(row['rooms_count']):
        if row['all_floor'] < 28:
            return 0  # Студия
        else:
            return row['rooms_count']
    else:
        return row['rooms_count']

df['rooms_count'] = df.apply(fill_room_count, axis=1)
```

```python
df['rooms_count'] = df.groupby(pd.cut(df['all_floor'], bins=5))['rooms_count'].transform(lambda x: x.fillna(x.median()))
```

### Раз не известно о типе санузла, то поставим нет.

```python
df['type_bathrooms'] = df['type_bathrooms'].fillna('Нет')
```

### Заполним жилую площадь по моде.

```python
df['live_squar'] = pd.to_numeric(df['live_squar'], errors='coerce')
live_value = df['live_squar'].mode()[0]
df['live_squar'] = df['live_squar'].fillna(live_value)
```

### Заполним площадь кухни по медиане. 

```python
df['kitchen_squar'] = pd.to_numeric(df['kitchen_squar'], errors='coerce')
kitchen_value = df['kitchen_squar'].mean()
df['kitchen_squar'] = df['kitchen_squar'].fillna(kitchen_value)
```

### Заполним тип ремонта по моде.

```python
repair_mode = df['repair'].mode()[0]
df['repair'] = df['repair'].fillna(repair_mode)
```

### Заполним высоту потолка по минимальному значению.

```python
min_value = df['ceil'].min()
df['ceil'] = df['ceil'].fillna(min_value)
```

### Заполним время до метро если его нет нулем.

```python
df['time_to_metro'] = df['time_to_metro'].fillna(0)
```

### в любом случае туалет и ванная есть в кв

```python
df['bathrooms'] = df['bathrooms'].fillna(1.0)
```

### Пусть машина сама заполнит типы комнат

```python
train_data = df[df['type_rooms'].notnull()]
X_train = train_data[['rooms_count']]
y_train = train_data['type_rooms']

model = RandomForestClassifier()
model.fit(X_train, y_train)

# Заполняем пропуски
missing_data = df[df['type_rooms'].isnull()]
X_missing = missing_data[['rooms_count']]
df.loc[df['type_rooms'].isnull(), 'type_rooms'] = model.predict(X_missing)
```
